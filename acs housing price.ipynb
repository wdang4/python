```python
def reduce_df(df,phrases_to_drop):
  df.columns=df.iloc[0]
  df = df.drop(index=[0])
  cols_to_drop=[]
  for col in df.columns:
    for phrase in phrases_to_drop:
      if phrase in col:
        cols_to_drop.append(col)
  df = df.drop(columns=cols_to_drop)
  return df
```

```python
drop_soc_phrases=['Margin of Error','ANCESTRY','YEAR OF ENTRY','REGION OF BIRTH OF FOREIGN BORN','LANGUAGE SPOKEN AT HOME',
                  'GRANDPARENTS','DISABILITY STATUS','RESIDENCE 1 YEAR AGO']
reduced_social_df = reduce_df(df,drop_soc_phrases)
reduced_social_df
```

```python
from sklearn_pandas import DataFrameMapper,FunctionTransformer
```

```python
def string_to_num(val):
    # if pd.isna(val):
    #     return pd.to_numeric(val, errors='coerce', downcast='float')
    # else:
    #     return float(val[1:])
    try:
        pd.to_numeric(val)
    except:
        return 0
    else:
        return pd.to_numeric(val)
```

```python
steps1 = [([col], None) for col in prototype_reduced_df.iloc[:,:2].columns]
steps2 = [([col], [FunctionTransformer(string_to_num)]) for col in prototype_reduced_df.iloc[:,2:].columns]
steps = steps1 + steps2
steps
```

```python
mapper = DataFrameMapper(steps,df_out=True)
mapper
```
